```python
def transformada( num_columna):
    """
Toma los datos obtenidos, le hace la transformada a los datos en la columna num_columna y devuelve
una matriz de amplitudes contra frecuencia
Input:

-num_columna: es el numero de la columna de los datos que se quiere transformar
Output:
transformada: Una matriz de [1,400] con n el numero de datos que es amplitud contra frecuencia
"""
    y=listofzeros = [0] * 400
    j=0

    for line in open("sampled+ma0844az_1-1+_data.txt"):
        columns = line.split( )
    
        if len(columns) >= 2:
            y[j]= columns[num_columna]
            j=j+1
        
    t = listofzeros = [0] * 400
    for i in range(0, 400):
        t[i]=i+1
    
    fft_x = fft(y) / n
    freq = fftfreq(n, 60)
    ab=np.abs(fft_x)
   

    

    return freq,fft_x



```

```python
def diezmayores( b):

    """
Toma una serie de datos y los vuelve todos cero excepto los 10 mayores
Input:
- b: los datos

Output:
b: Una lista con solo los 10 datos mayores y el resto cero
"""
    ab=(np.abs(b))**2
    orde=sort(ab)
    for x in range(0, 400):
        if(ab[x]<orde[390]):
            b[x]=0
    return b
```

```python
def unacolumna (num_columna):

    """
Devuelve dolo una columna de los datos del archivo
Input:
- num_columna: El numero de la columna que queremos

Output:
y: Una lista con la columna que queriamos
"""
    y=listofzeros = [0] * 400
    j=0
    for line in open("sampled+ma0844az_1-1+_data.txt"):
        columns = line.split( )
    
        if len(columns) >= 2:
            y[j]= float(columns[num_columna])
            j=j+1
    return y
```

```python
#FUNCIONES
def ultimo_99(datosiniciales):
    """
    Busca en el archivo datosiniciales el valor -99 y guarda la informaci'on despues del registro.
    cuando encuentra el ultimo -99 guarda todos los datos desde ah'i en adelante
    
    Input:Matriz de datosiniciales
    Output:Matriz de datos sin los valores de -99
    """
    datoslimpios=np.ones((6,4)) #Depende de la fila en la que se encuentre el ultimo -99
    
    return datoslimpios

def filtrar(datpfilt):
    """
    Con esta funcion se pretende tener la mejor medicion de manchas por cada a-no (eliminando los meses menos significativos)
    Analiza todas las columnas de la matriz de entrada.
    Las columnas 2 y la 3 para tener el valor de manchas ,obtenido por mayor tiempo de observacion en el a-no especifico, dado por la columna 1.
    
    Input datpfilt: Es la matriz de datos que tiene informacion sobre algunos meses y los dias que se observo en cada mes. Datos para filtrar

    Output tfiltrado: Es el vector de tiempo (no equispaciado) de solo una medida por a-no
    Output numMfilt: Es el vector de numero de manchas registradas con un solo dato por a-no
    """
    tfiltrado=datpfilt[:,0] #El tama-no depende de los valores que se tenga en datos por ahora inicializar asi. Es necesario usar append
    numMfilt=datpfilt[:,1] #El tama-no depende de los valores que se tenga en datos por ahora inicializar asi. Es necesario hacer append porque no se conoce el tama;o
    
    return tfiltrado, numMfilt

def t_completo(a,b):
    """
    Crea vector de tiempo completo equiespaciado a 1 a-no. Usamos  linspace()

    Input a : A-no de inicio de las medidas.
    Input b : A-no de la ultima toma de los datos.

    Output tcompleto : vector de tiempo 
    """
    tcompleto=np.linspace(a,b,1)
    return tcompleto

#INTERPOLACIONES,
def interpolaciones(t_dato,mancha,grado):
    
    """
    Realiza la interpolacion usando la funcion de scipy

    Input t_dato: es el vector de tiempo
    Input mancha: es el vector de num de manchas
    Input grado : es el grado de la interpolacion que quiere si es 0 cte, 1 lineal, 3 cubica.
    
    Output f: La funcion que interpola dependiendo el grado

    """
    f=interp1d(t_dato,mancha,grado)
    return f

#Transformada de fourier
def tfourier(tm,fs,n):
    """
    Calcula la transformada de fourier y los coeficientes de fourier correspondientes

    Input tm:el vector de tiempo equiespaciado
    Input fs: el valor de las manchas con la respectiva interpolacion

    Output fft,frec
    """
    trans=fs # Se usa la funcion fft()
    frec = fftfreq(n, 1) 
    return trans,frec
def hacer0(frec, ampfft):
    """
    Esta funcion elimina los valores de frecuencia que no se encuentran en el rango [1/20,1/2]

    Input frec: Es el array de frecencia dada en la transformada
    Input ampfft: Es el array de peso para la frec. dada en la transformada

    Output fre: Es el array limpio de frecuencias fuera del rango
    Output amp: Amplitudes que sobrevivieron
    """
    fre=frec
    amp=ampfft
    return fre, amp

def transinversa(vfrec, vamp):
    """
    A partir de los valores encontrados al limpiar la se;al se reconstruye la grafica con los valores de frecuencia significativos
    Esto se hace por medio de la transformada inversa

    Input vfrec: Es el array de frecencia dada despues de la limpieza
    Input vamp: Es el array de peso para la frec. despues de la limpieza

    Output tiempo: Es el array el el espacio del tiempo, o la aproximacion de la se;al inicial
    Output nummanchas: es el numero de manchas para cada tiempo.
    """
    tiempo=vfrec#TEMPORAL
    nummanchas=vamp#TEMPORAL
    return tiempo, nummanchas

def periodo(tf0,numf0,tf1,numf1):#Tambien el 3
    """
    Esta funcion analiza los datos y presenta la solucion del periodo del ciclo solar. Analiza cada uno de los periodos compara

    Input tf0,1,3, numf0,1,3: son las se;ales reconstruidas con las 3 diferentes interpolaciones.

    Otput p: Periodo en a;os.
    """
    p=10#TEMPORAL
    return p

def normalizar(t0,m0,t1,m1,t3,m3):
    """
    Calcula lo log para normalizar los datos

    INPUTS: Se;ales 
    Output:Se;ales normalizadas
    """
    tn0=t0
    mn0=m0
    tn1=t1
    mn1=m1
    tn3=t3
    mn3=m3
    
    return tn0,mn0,tn1,mn1, tn3,mn3
```

```python
%pylab inline
import numpy as np
from scipy.interpolate import interp1d
from scipy.fftpack import fft, fftfreq
import matplotlib.pyplot as plt

datosiniciales=np.loadtxt(open('sparse_sample_monthrg.dat','r'))
#Desde la fila 252 del archivo no aparecen mas -99 se realizo el calculo a mano.
datosnuevos=datosiniciales[252:,:]
nfilas=datosnuevos.shape[0]
ai=datosnuevos[0,0]#a~no de inicio
af=datosnuevos[nfilas-1,0]#a~no final
mi=datosnuevos[0,1]#mes inicial
mf=datosnuevos[nfilas-1,1]#mes final
# Numero de puntos totales equiespaciados a un mes
puntos=((af-ai+1)*12)-mi+1-(12-mf)

colum1=datosnuevos[:,0]#array de a~nos
colum2=datosnuevos[:,1]#array de meses
manchas=datosnuevos[:,3]#array de numero de manchas prom
r=len(colum1)#numero de datos tomados
t=np.zeros(r)
tcompleto=np.linspace(0,puntos-1,puntos)#Vector de tiempo equiespaciadopara poder hacer la interpolacion

#INICIALIZAR t de los datos en terminos de meses
for i in range(r):
    a=colum1[i]
    m=colum2[i]
    pos=((a-ai)*12)+m-mi
    t[i]=pos  
    
#Funciones INTERPOLACIONES
f0=interp1d(t,manchas,0)
f1=interp1d(t,manchas,1)
f3=interp1d(t,manchas,3)
manchas0=f0(tcompleto)#Interpoacion constqnt
manchas1=f1(tcompleto)#Interpolacion lineal
manchas3=f3(tcompleto)#Interpolacion cubica

n=len(manchas0)


#GRAFICAS DE LAS INTERPOLACIONES
scatter(t,manchas,s=10)
plot(tcompleto, manchas0)#grafica Interpoacion constqnt
plot(tcompleto, manchas1)#grafica Interpolacion lineal
plot(tcompleto, manchas3)#grafica Interpolacion cubica
legend(['const', 'linear','cubica', 'data'])

plt.suptitle('INTERPOLACIONES')
plt.xlabel("Tiempo[meses]")
plt.ylabel("Manchas promedio")
```

```python
#TRANSFORMADAS
fft_x0=fft(manchas0)/n# Transformada normalizada para int.cte
freq0=fftfreq(n,1)#Frecuencias de la transformada para int.cte
plt.plot(freq0, np.abs(fft_x0))
```

```python
fft_x1=fft(manchas1)/n# Transformada normalizada para int.lin
freq1=fftfreq(n,1)#Frecuencias de la transformada para int.lin
plt.plot(freq1, np.abs(fft_x1))
```

```python
fft_x3=fft(manchas3)/n# Transformada normalizada para int.CUB
freq3=fftfreq(n,1)#Frecuencias de la transformada para int.CUB
plt.plot(freq3, np.abs(fft_x3))
```

```python
plt.plot(freq0, np.abs(fft_x0))
plt.plot(freq1, np.abs(fft_x1))
plt.plot(freq3, np.abs(fft_x3))
legend(['const', 'linear','cubica'])
plt.suptitle('Espectros')
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
```

```python
fft_x0_shifted=np.fft.fftshift(fft_x0)
freq0_shifted=np.fft.fftshift(freq0)
plt.plot(freq0_shifted, np.abs(fft_x0_shifted))

fft_x1_shifted=np.fft.fftshift(fft_x1)
freq1_shifted=np.fft.fftshift(freq1)
plt.plot(freq1_shifted, np.abs(fft_x1_shifted))

fft_x3_shifted=np.fft.fftshift(fft_x3)
freq3_shifted=np.fft.fftshift(freq3)
plt.plot(freq3_shifted, np.abs(fft_x3_shifted))
```

```python
half_n = np.ceil(n/2.0)

fft_x0_half = (2.0) * fft_x0[:half_n]
freq0_half = freq0[:half_n]
plt.plot(freq0_half, np.abs(fft_x0_half))

fft_x1_half = (2.0) * fft_x1[:half_n]
freq1_half = freq1[:half_n]
plt.plot(freq1_half, np.abs(fft_x1_half))

fft_x3_half = (2.0) * fft_x3[:half_n]
freq3_half = freq3[:half_n]
plt.plot(freq3_half, np.abs(fft_x3_half))

#print amp0, amp1, amp3
```

```python
#LIMPIAR LOS DATOS volver cero los valores que no se encuentran entre el rango de 2 a 20 a;os#Vectores de frecuencia

#vector de frecuenciass
fre0=freq0
fre1=freq1
fre3=freq3
#Vectores de amplitud
amp0=fft_x0
amp1=fft_x1
amp3=fft_x3

#Frecuencias limite
fd=4.16666e-3#frecuencia corresponciente a periodo mayor a a;os
fu=0.041666#frecuencia correspondiente a periodo menor a 2 a;os

amp0[np.abs(fre0)>fu]=0
amp0[np.abs(fre0)<fd]=0
amp1[np.abs(fre1)>fu]=0
amp1[np.abs(fre1)<fd]=0
amp3[np.abs(fre3)>fu]=0
amp3[np.abs(fre3)<fd]=0

#GRAFICAS espectro de frecuencias filtrado

plot(fre0,np.abs(amp0))
plot(fre1, np.abs(amp1))
plot(fre3, np.abs(amp3))
xlim([fd,fu])
legend(['const', 'linear','cubica'])
plt.suptitle('Espectros filtrado')
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
```

```python
#Transformada inversa para las funciones filtradas
flimpia0=n*ifft(amp0)#Transformada inversa cte
flimpia1=n*ifft(amp1)#Transformada inversa lin
flimpia3=n*ifft(amp3)#Transformada inversa cubic

#GRAFICA RECONSTRUCCION DESPUES DE FILTRAR y Normalizada
plot(tcompleto/12,real(flimpia0))
plot(tcompleto/12,real(flimpia1))
plot(tcompleto/12,real(flimpia3))


legend(['const', 'linear','cubica','datos'])
plt.suptitle('Numero de manchas')
plt.xlabel("Tiempo(meses)")
plt.ylabel("Manchas promedio")
    
```
